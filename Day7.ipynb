```python
import pandas as pd

data = {
    'Numerical Data': [10, 25.5, -5, 1000],
    'Categorical Data': ['Red', 'Blue', 'Red', 'Green'],
    'Text Data': ['This is a sentence.', 'Another example.', 'Short text.', 'A very long piece of text to illustrate.']
}

df = pd.DataFrame(data)

print("DataFrame:")
display(df)

print("\nData Types:")
display(df.dtypes)

print("\nOperations:")

print("Mean of Numerical Data:", df['Numerical Data'].mean())

print("Value Counts of Categorical Data:\n", df['Categorical Data'].value_counts())

print("Length of Text Data:\n", df['Text Data'].str.len())
```
