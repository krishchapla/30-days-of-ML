# Task
Preprocess the data in the file "grid.xlsx" by handling missing values, applying one-hot encoding to categorical features, and scaling numerical features. Display the first 5 rows of the preprocessed data.

## Handle missing values

### Subtask:
Identify and handle missing values in the DataFrame.


```python

df = pd.read_excel("/Market analysis.xlsx")
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
print("Columns with missing values and their counts:")
print(missing_values)
```

```python
numerical_cols_with_missing = missing_values.index.tolist()
for col in numerical_cols_with_missing:
    df[col] = df[col].fillna(df[col].mean())

print("\nMissing values after filling:")
print(df.isnull().sum())
```

## Encode categorical variables

### Subtask:
Identify categorical columns and apply one-hot encoding.


```python
categorical_cols = df.select_dtypes(include='object').columns
if len(categorical_cols) > 0:
    df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
df.head()
```

## Scale features

### Subtask:
Identify numerical columns and apply feature scaling.


```python
numerical_cols = df.select_dtypes(include=np.number).columns
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```
