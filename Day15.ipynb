```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/drive/MyDrive/Datasets/zoo (1).csv')
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

y_true = df['animal'].apply(lambda x: 0 if x == 'elephant' else (1 if x == 'tiger' else -1)).tolist()
# Let's assume a simple prediction: if water_need is > 400, predict elephant (0), otherwise predict tiger (1)
y_pred = df['water_need'].apply(lambda x: 0 if x > 400 else 1).tolist()

# Filter out rows where animal is neither elephant nor tiger, as they are not included in our simple classification
valid_indices = [i for i, label in enumerate(y_true) if label != -1]
y_true = [y_true[i] for i in valid_indices]
y_pred = [y_pred[i] for i in valid_indices]

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
