First, let's create a couple of NumPy arrays.

```python
import numpy as np

array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([6, 7, 8, 9, 10])

print("Array 1:", array1)
print("Array 2:", array2)
```

Now, let's perform some basic arithmetic operations on these arrays.

```python
# Addition
sum_array = array1 + array2
print("Sum of arrays:", sum_array)

# Subtraction
diff_array = array2 - array1
print("Difference of arrays:", diff_array)

# Multiplication
prod_array = array1 * array2
print("Product of arrays:", prod_array)

# Division
div_array = array2 / array1
print("Division of arrays:", div_array)
```

You can also perform scalar operations (operations with a single number).

```python
# Add a scalar
array1_plus_scalar = array1 + 10
print("Array 1 + 10:", array1_plus_scalar)

# Multiply by a scalar
array2_times_scalar = array2 * 2
print("Array 2 * 2:", array2_times_scalar)
```

NumPy also provides many built-in functions for array operations, such as finding the sum, mean, and standard deviation.

```python
# Sum of all elements in an array
total_sum = np.sum(array1)
print("Sum of elements in Array 1:", total_sum)

# Mean of elements in an array
mean_value = np.mean(array2)
print("Mean of elements in Array 2:", mean_value)

# Standard deviation of elements in an array
std_dev = np.std(array1)
print("Standard deviation of Array 1:", std_dev)
```
